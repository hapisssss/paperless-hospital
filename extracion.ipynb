````python
import os
import re
import unicodedata

def sanitize_filename(name, max_length=120):
    """
    Convert any string into a filesystem-safe filename.
    """

    # Normalize unicode (avoid weird hidden characters)
    name = unicodedata.normalize("NFKD", str(name))

    # Remove control characters
    name = re.sub(r'[\x00-\x1f\x7f]', '', name)

    # Remove invalid Windows characters
    name = re.sub(r'[\\/*?:"<>|]', '', name)

    # Remove leading/trailing whitespace
    name = name.strip()

    # Remove trailing dots (Windows restriction)
    name = name.rstrip(".")

    # Prevent empty filename
    if not name:
        name = "unknown"

    # Prevent reserved Windows names
    reserved = {
        "CON","PRN","AUX","NUL",
        "COM1","COM2","COM3","COM4","COM5","COM6","COM7","COM8","COM9",
        "LPT1","LPT2","LPT3","LPT4","LPT5","LPT6","LPT7","LPT8","LPT9"
    }

    if name.upper() in reserved:
        name = f"{name}_file"

    # Limit length (prevent path > 255)
    name = name[:max_length]

    return name
````

````python
import os
import re
				
# Extraction Aspek Koding Documents
def extraction_aspek_koding(data):
    # Make folder if doesn't exists
    output_dir = "Knowledge"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Original logic (unchanged)
    raw_file_name = data["Prosedur"] if data["Diagnosis"] == "-" else data["Diagnosis"]

    # Sanitize filename
    safe_name = sanitize_filename(raw_file_name)

    file_path = os.path.join(output_dir, f"{safe_name}.md")

    # Prevent accidental overwrite
    counter = 1
    base_path = file_path
    while os.path.exists(file_path):
        file_path = os.path.join(output_dir, f"{safe_name}_{counter}.md")
        counter += 1

    # Document Template
    document = f"""

**DOMAIN:**
{data['Domain']}

**KATEGORI ICD:**
{data['Kategori ICD']}

**DIAGNOSIS:**
{data["Diagnosis"]}

**PROSEDUR:**
{data["Prosedur"]}

**ASPEK KODING:**
{data["Aspek Koding"]}

**PERHATIAN KHUSUS:**
{data["Perhatian Khusus"]}
"""

    # Read File
    try:
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(document)
        return "success"
    except Exception as error:
        print("Failed:", file_path)
        print("Error:", error)
        return "failed"

	
# Extraction Aspek Medis Documents
def extraction_aspek_medis(data):
    # Make folder if doesn't exists
    output_dir = "Knowledge"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Original logic (unchanged)
    raw_file_name = data["Prosedur"] if data["Diagnosis"] == "-" else data["Diagnosis"]

    # Sanitize filename
    safe_name = sanitize_filename(raw_file_name)

    file_path = os.path.join(output_dir, f"{safe_name}.md")

    # Prevent accidental overwrite
    counter = 1
    base_path = file_path
    while os.path.exists(file_path):
        file_path = os.path.join(output_dir, f"{safe_name}_{counter}.md")
        counter += 1

    # Document Template
    document = f"""


**DIAGNOSIS:**
{data["Diagnosis"]}

**PROSEDUR:**
{data["Prosedur"]}

**ASPEK MEDIS:**
{data["Aspek Medis"]}

**PERHATIAN KHUSUS:**
{data["Perhatian Khusus"]}
"""

    # Read File
    try:
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(document)
        return "success"
    except Exception as error:
        print("Failed:", file_path)
        print("Error:", error)
        return "failed"


# Extraction Aspek Administrasi Documents
def extraction_aspek_administrasi(data):
    
    # Make folder if doesn't exists
    output_dir = "Knowledge"
    os.makedirs(output_dir, exist_ok=True)

    # Original logic (unchanged)
    if data["Diagnosis"] == "-":
        raw_file_name = data["Prosedur"]
    elif data["Prosedur"] == "-":
        raw_file_name = data["Perihal Administrasi"]
    else:
        raw_file_name = data["Diagnosis"]

    # Use the SAME sanitizer
    safe_name = sanitize_filename(raw_file_name)

    file_path = os.path.join(output_dir, f"{safe_name}.md")

    # Prevent overwrite
    counter = 1
    while os.path.exists(file_path):
        file_path = os.path.join(output_dir, f"{safe_name}_{counter}.md")
        counter += 1

    document = f"""

**DIAGNOSIS:**
{data["Diagnosis"]}

**PROSEDUR:**
{data["Prosedur"]}

**PERIHAL ADMINISTRASI:**
{data["Perihal Administrasi"]}

**SOLUSI ADMINISTRASI:**
{data["Solusi Administrasi"]}

**PERHATIAN KHUSUS:**
{data["Perhatian Khusus"]}
"""

    try:
        with open(file_path, "w", encoding="utf-8") as file:
            file.write(document)
        return "success"
    except Exception as error:
        print("Failed:", file_path)
        print("Error:", error)
        return "failed"
````
